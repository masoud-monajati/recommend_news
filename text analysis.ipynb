# NLP on titles

```python
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
```

```python
# combining different stop words.
stop_words_skt = text.ENGLISH_STOP_WORDS

stop_words_en = stopwords.words('english')

print("# of stop_words_skt:\t\t %s" % len(stop_words_skt))
print("# of stop_words_en:\t\t %s" % len(stop_words_en))
print("# of punctuation:\t\t %s" % len(punctuation))
combined_stopwords = set.union(set(stop_words_en),set(punctuation),set(stop_words_skt))
print("# of combined_stopwords:\t %s" % len(combined_stopwords))
```

```python

wnl = nltk.wordnet.WordNetLemmatizer()
analyzer = CountVectorizer().build_analyzer()

def penn2morphy(penntag):
    """ Converts Penn Treebank tags to WordNet. """
    morphy_tag = {'NN':'n', 'JJ':'a',
                  'VB':'v', 'RB':'r'}
    try:
        return morphy_tag[penntag[:2]]
    except:
        return 'n'
    
def lemmatization(list_word):
    return [wnl.lemmatize(word.lower(), pos=penn2morphy(tag)) 
            for word, tag in pos_tag(list_word)]

def stem_rmv_punc(doc):
    return (word for word in lemmatization(analyzer(doc)) if word not in combined_stopwords and not word.isdigit())
```

```python
count_vect1 = CountVectorizer(min_df=3,analyzer=stem_rmv_punc, stop_words='english')
train_matrix = count_vect1.fit_transform(train_data)
trst_matrix = count_vect1.transform(trst_data)
```

```python
print ("Shape of the document-term matrix of the train set is: ", train_matrix.shape)
print ("Shape of the document-term matrix of the test set is: ", trst_matrix.shape)
```

```python
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(train_matrix)
X_trst_tfidf = tfidf_transformer.transform(trst_matrix)
print ("Shape of the TF-IDF matrix of the train set is: ", X_train_tfidf.shape)
print ("Shape of the TF-IDF matrix of the test set is: ", X_trst_tfidf.shape)
```

# dimensionally reduction (LSI)

```python
#LSI
from sklearn.decomposition import TruncatedSVD

LSI_SVD = TruncatedSVD(n_components=80)
LSI_eight_train = LSI_SVD.fit_transform(X_train_tfidf)
LSI_eight_trst = LSI_SVD.transform(X_trst_tfidf)
print("Shape of the LSI-reduced TF-IDF matrix of the train set is: ", LSI_eight_train.shape)
print("Shape of the LSI-reduced TF-IDF matrix of the test set is: ", LSI_eight_trst.shape)
```

# dimensionally reduction (NMF)

```python
#NMF
from sklearn.decomposition import NMF
NMF_model = NMF(n_components=80, init='random')
W_train = NMF_model.fit_transform(X_train_tfidf)
W_trst = NMF_model.transform(X_trst_tfidf)
print("Shape of the NMF-reduced TF-IDF matrix of the train set is: ", W_train.shape)
print("Shape of the NMF-reduced TF-IDF matrix of the test set is: ", W_trst.shape)
```

```python
def plot_confusion_matrix(cm, title='Confusion matrix', cmap=plt.cm.Blues,n=4):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(n)
    plt.xticks(tick_marks, rotation=45)
    plt.yticks(tick_marks)
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```
